```python
df= df.drop('Name',axis=1)
```

```python
Genre_Average_Rating = df.groupby('Genre')['Rating'].transform('mean')
df['Genre_Average_Rating'] =Genre_Average_Rating
Director_Average_Rating = df.groupby('Director')['Rating'].transform('mean')
df['Director_Average_Rating'] =Director_Average_Rating
Actor1_Average_Rating = df.groupby('Actor 1')['Rating'].transform('mean')
df['Actor1_Average_Rating'] =Actor1_Average_Rating
Actor2_Average_Rating = df.groupby('Actor 2')['Rating'].transform('mean')
df['Actor2_Average_Rating'] =Actor2_Average_Rating
Actor3_Average_Rating = df.groupby('Actor 3')['Rating'].transform('mean')
df['Actor3_Average_Rating'] =Actor3_Average_Rating
df
```

```python
from sklearn.model_selection import train_test_split
x= df[['Year','Votes','Duration','Genre_Average_Rating','Director_Average_Rating','Actor1_Average_Rating','Actor2_Average_Rating','Actor3_Average_Rating']]
y=df['Rating']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
train_data= x_train.join(y_train)
train_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, StackingRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
import lightgbm as lgb
from ngboost import NGBRegressor

# Assuming train_data and test_data are already defined

x_train, y_train = train_data[['Year','Votes','Duration','Genre_Average_Rating','Director_Average_Rating','Actor1_Average_Rating','Actor2_Average_Rating','Actor3_Average_Rating']], train_data['Rating']

# Linear Regression
fitted_model_lr = LinearRegression()
fitted_model_lr.fit(x_train,y_train)
y_pred_lr = fitted_model_lr.predict(x_test)

# Random Forest
fitted_model_rf = RandomForestRegressor()
fitted_model_rf.fit(x_train, y_train)
y_pred_rf = fitted_model_rf.predict(x_test)

# Gradient Boosting
fitted_model_gb = GradientBoostingRegressor()
fitted_model_gb.fit(x_train, y_train)
y_pred_gb = fitted_model_gb.predict(x_test)

# XGBoost
fitted_model_xgb = XGBRegressor()
fitted_model_xgb.fit(x_train, y_train)
y_pred_xgb = fitted_model_xgb.predict(x_test)

# CatBoost
fitted_model_cb = CatBoostRegressor()
fitted_model_cb.fit(x_train, y_train)
y_pred_cb = fitted_model_cb.predict(x_test)

# LightGBM
fitted_model_lgb = lgb.LGBMRegressor()
fitted_model_lgb.fit(x_train, y_train)
y_pred_lgb = fitted_model_lgb.predict(x_test)

# AdaBoost
fitted_model_ab = AdaBoostRegressor()
fitted_model_ab.fit(x_train, y_train)
y_pred_ab = fitted_model_ab.predict(x_test)

# NGBoost
fitted_model_ngb = NGBRegressor()
fitted_model_ngb.fit(x_train, y_train)
y_pred_ngb = fitted_model_ngb.predict(x_test)

# Stacking Regressor
estimators = [('lr', LinearRegression()), ('rf', RandomForestRegressor()), ('gb', GradientBoostingRegressor())]
fitted_model_sr = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())
fitted_model_sr.fit(x_train, y_train)
y_pred_sr = fitted_model_sr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score#

# Linear Regression
MSE_lr = mean_squared_error(y_test, y_pred_lr)
MAE_lr = mean_absolute_error(y_test, y_pred_lr)
R2_Score_lr = r2_score(y_test, y_pred_lr)
MAPE_lr = MAE_lr / np.mean(y_test)  # MAPE formula
Accuracy_lr = (1 - MAPE_lr) * 100  # Convert to percentage

print('   Performance Evaluation for Linear Regression Model: ')
print('Mean Squared Error (MSE): ', MSE_lr)
print('Mean Absolute Error (MAE): ', MAE_lr)
print('R² Score: ', R2_Score_lr)
print(f'Accuracy Percentage: {Accuracy_lr:.2f}%')

# Random Forest Model Evaluation
MSE_rf = mean_squared_error(y_test, y_pred_rf)
MAE_rf = mean_absolute_error(y_test, y_pred_rf)
R2_Score_rf = r2_score(y_test, y_pred_rf)
MAPE_rf = MAE_rf / np.mean(y_test)  # MAPE formula
Accuracy_rf = (1 - MAPE_rf) * 100  # Convert to percentage

print('\n   Performance Evaluation for Random Forest Model: ')
print('Mean Squared Error (MSE): ', MSE_rf)
print('Mean Absolute Error (MAE): ', MAE_rf)
print('R² Score: ', R2_Score_rf)
print(f'Accuracy Percentage: {Accuracy_rf:.2f}%')

# Gradient Boosting Model Evaluation
MSE_gb = mean_squared_error(y_test, y_pred_gb)
MAE_gb = mean_absolute_error(y_test, y_pred_gb)
R2_Score_gb = r2_score(y_test, y_pred_gb)
MAPE_gb = MAE_gb / np.mean(y_test)  # MAPE formula
Accuracy_gb = (1 - MAPE_gb) * 100  # Convert to percentage

print('\n   Performance Evaluation for Gradient Boosting Model: ')
print('Mean Squared Error (MSE): ', MSE_gb)
print('Mean Absolute Error (MAE): ', MAE_gb)
print('R² Score: ', R2_Score_gb)
print(f'Accuracy Percentage : {Accuracy_gb:.2f}%')

# XGBoost Model Evaluation
MSE_xgb = mean_squared_error(y_test, y_pred_xgb)
MAE_xgb = mean_absolute_error(y_test, y_pred_xgb)
R2_Score_xgb = r2_score(y_test, y_pred_xgb)
MAPE_xgb = MAE_xgb / np.mean(y_test)  # MAPE formula
Accuracy_xgb = (1 - MAPE_xgb) * 100  # Convert to percentage

print('\n   Performance Evaluation for XGBoost Model: ')
print('Mean Squared Error (MSE): ', MSE_xgb)
print('Mean Absolute Error (MAE): ', MAE_xgb)
print('R² Score: ', R2_Score_xgb)
print(f'Accuracy Percentage: {Accuracy_xgb:.2f}%')

# CatBoost Model Evaluation
MSE_cb = mean_squared_error(y_test, y_pred_cb)
MAE_cb = mean_absolute_error(y_test, y_pred_cb)
R2_Score_cb = r2_score(y_test, y_pred_cb)
MAPE_cb = MAE_cb / np.mean(y_test)  # MAPE formula
Accuracy_cb = (1 - MAPE_cb) * 100  # Convert to percentage

print('\n   Performance Evaluation for CatBoost Model: ')
print('Mean Squared Error (MSE): ', MSE_cb)
print('Mean Absolute Error (MAE): ', MAE_cb)
print('R² Score: ', R2_Score_cb)
print(f'Accuracy Percentage: {Accuracy_cb:.2f}%')

# AdaBoost Model Evaluation
MSE_ab = mean_squared_error(y_test, y_pred_ab)
MAE_ab = mean_absolute_error(y_test, y_pred_ab)
R2_Score_ab = r2_score(y_test, y_pred_ab)
MAPE_ab = MAE_ab / np.mean(y_test)  # MAPE formula
Accuracy_ab = (1 - MAPE_ab) * 100  # Convert to percentage

print('\n   Performance Evaluation for AdaBoost Model: ')
print('Mean Squared Error (MSE): ', MSE_ab)
print('Mean Absolute Error (MAE): ', MAE_ab)
print('R² Score: ', R2_Score_ab)
print(f'Accuracy Percentage: {Accuracy_ab:.2f}%')

# LightGBM Model Evaluation
MSE_lgb = mean_squared_error(y_test, y_pred_lgb)
MAE_lgb = mean_absolute_error(y_test, y_pred_lgb)
R2_Score_lgb = r2_score(y_test, y_pred_lgb)
MAPE_lgb = MAE_lgb / np.mean(y_test)  # MAPE formula
Accuracy_lgb = (1 - MAPE_lgb) * 100  # Convert to percentage

print('\n   Performance Evaluation for LightGBM Model: ')
print('Mean Squared Error (MSE): ', MSE_lgb)
print('Mean Absolute Error (MAE): ', MAE_lgb)
print('R² Score: ', R2_Score_lgb)
print(f'Accuracy Percentage: {Accuracy_lgb:.2f}%')

# NGBoost Model Evaluation
MSE_ngb = mean_squared_error(y_test, y_pred_ngb)
MAE_ngb = mean_absolute_error(y_test, y_pred_ngb)
R2_Score_ngb = r2_score(y_test, y_pred_ngb)
MAPE_ngb = MAE_ngb / np.mean(y_test)  # MAPE formula
Accuracy_ngb = (1 - MAPE_ngb) * 100  # Convert to percentage

print('\n   Performance Evaluation for NGBoost Model: ')
print('Mean Squared Error (MSE): ', MSE_ngb)
print('Mean Absolute Error (MAE): ', MAE_ngb)
print('R² Score: ', R2_Score_ngb)
print(f'Accuracy Percentage: {Accuracy_ngb:.2f}%')

# Stacking Regressor Model Evaluation
MSE_sr = mean_squared_error(y_test, y_pred_sr)
MAE_sr = mean_absolute_error(y_test, y_pred_sr)
R2_Score_sr = r2_score(y_test, y_pred_sr)
MAPE_sr = MAE_sr / np.mean(y_test)  # MAPE formula
Accuracy_sr = (1 - MAPE_sr) * 100  # Convert to percentage

print('\n   Performance Evaluation for Stacking Regressor Model: ')
print('Mean Squared Error (MSE): ', MSE_sr)
print('Mean Absolute Error (MAE): ', MAE_sr)
print('R² Score: ', R2_Score_sr)
print(f'Accuracy Percentage : {Accuracy_sr:.2f}%')
```

```python
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest', 'Gradient Boosting', 'XGBoost', 'CatBoost', 'AdaBoost', 'LightGBM', 'NGBoost', 'Stacking Regressor'],
    'Mean Squared Error (MSE)': [MSE_lr,MSE_rf, MSE_gb, MSE_xgb, MSE_cb, MSE_ab, MSE_lgb, MSE_ngb, MSE_sr ],
    'R-squared score (R2)': [R2_Score_lr,R2_Score_rf,R2_Score_gb, R2_Score_xgb, R2_Score_cb, R2_Score_ab, R2_Score_lgb, R2_Score_ngb, R2_Score_sr ],
    'Accuracy ': [Accuracy_lr, Accuracy_rf, Accuracy_gb, Accuracy_xgb, Accuracy_cb, Accuracy_ab, Accuracy_lgb, Accuracy_ngb, Accuracy_sr]
})
print(results)
```

```python
models = results['Model']
mse_scores = results['Mean Squared Error (MSE)'].round(2)
r2_scores = results['R-squared score (R2)'].round(2)
num_models = len(models)
bar_width = 0.25
index = np.arange(num_models)
```

```python
plt.figure(figsize=(12, 8))

plt.bar(index, mse_scores, bar_width, label='Mean Squared Error (MSE)', color='pink', alpha=0.7)
plt.bar(index + bar_width, r2_scores, bar_width, label='R-squared (R2)', color='purple', alpha=0.7)

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Comparison of MSE, R-squared (R2), Accuracy Scores of different models')
plt.xticks(index + bar_width, models, rotation=45)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
